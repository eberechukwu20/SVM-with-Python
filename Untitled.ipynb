```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,Y_train,Y_test = train_test_split(X,Y,test_size=0.2,random_state=101)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
len(X_train)
```

```python
from sklearn.svm import SVC
```

```python
machine = SVC()
```

```python
machine.fit(X_train,Y_train)
```

```python
mach_pred = machine.predict(X_test)
```

# Evaluating our model performance

```python
machine.score(X_test,Y_test)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
confusion_matrix(Y_test,mach_pred)
```

```python
sns.heatmap(confusion_matrix(Y_test,mach_pred),annot = True);
```

```python
print(classification_report(Y_test,mach_pred))
```
